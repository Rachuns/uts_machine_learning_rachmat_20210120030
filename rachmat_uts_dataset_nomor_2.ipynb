```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/marsani/UTS/main/dataset_uts-2021.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
df.info()
```

```python
# b. Menghapus Baris yang hilang (Ada Nilai NaN nya)
df.dropna(inplace=True)
df
```

```python
# c. Menghapus kolom yang berisi data yang tidak relevan (Hapus Kolom No)
df.drop(columns=['No'], inplace=True)
df
```

```python
# d. Perbaiki kesalahan input (intenal harusnya Internal)
df['Project'] = df['Project'].replace('intenal', 'Internal')
df
```

```python
# a. Menghapus Baris yang berulang Versi 2 Setelah Update Project intenal to Internal
df.drop_duplicates(inplace=True)
print(df)
duplicate_rows = df[df.duplicated()]
print("\nBaris yang sudah terhapus:")
print(duplicate_rows)
```

```python
# e. Ubah Tipe data float64 untuk Mobile Phone menjadi Teks atau Object
df['Mobile Phone'] = df['Mobile Phone'].astype(str)
df
```

```python
# f. Ganti nama semua kolom dalam bahasa Indonesia
df.columns = ['Nama', 'Alamat', 'Telepon Mobile', 'Projek', 'Email']
df
```

```python
# g. Simpan dalam file dataset yang baru
df.to_csv('dataset_cleaned.csv', index=False)
```
